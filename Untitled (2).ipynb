```python
import string
def quitar(linea):
    for c in string.punctuation:
        linea= linea.replace(c,"")
    return linea
```

```python
je = k.flatMap(f)
```

```python
je = je.map(quitar)
```

```python
tt = je.map(lambda x: (x,1))
tt.take(3)
```

```python
def gg(x,y):
    return x + y
```

```python
jhgj = tt.reduceByKey(gg)
```

```python
l = jhgj.collect()
```

```python
len(l)
```

```python
def r(x):
    return x[1]
```
